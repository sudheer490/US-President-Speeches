```python
import nltk
from collections import defaultdict
from nltk.corpus import inaugural
import pandas as pd

# Ensure the inaugural corpus is downloaded
nltk.download('inaugural')
nltk.download('stopwords')

# Prepare a dictionary for party affiliations based on external research
party_affiliations = {
    '1789-Washington': 'Other', '1793-Washington': 'Other', '1797-Adams': 'Other', 
    '1801-Jefferson': 'Other', '1805-Jefferson': 'Other', '1809-Madison': 'Other',
    '1813-Madison': 'Other', '1817-Monroe': 'Other', '1821-Monroe': 'Other',
    '1825-Adams': 'Other', '1829-Jackson': 'Democrat', '1833-Jackson': 'Democrat',
    '1837-VanBuren': 'Democrat', '1841-Harrison': 'Other', '1845-Polk': 'Democrat',
    '1849-Taylor': 'Other', '1853-Pierce': 'Democrat', '1857-Buchanan': 'Democrat',
    '1861-Lincoln': 'Republican', '1865-Lincoln': 'Republican', '1869-Grant': 'Republican',
    '1873-Grant': 'Republican', '1877-Hayes': 'Republican', '1881-Garfield': 'Republican',
    '1885-Cleveland': 'Democrat', '1889-Harrison': 'Republican', '1893-Cleveland': 'Democrat',
    '1897-McKinley': 'Republican', '1901-McKinley': 'Republican', '1905-Roosevelt': 'Republican',
    '1909-Taft': 'Republican', '1913-Wilson': 'Democrat', '1917-Wilson': 'Democrat',
    '1921-Harding': 'Republican', '1925-Coolidge': 'Republican', '1929-Hoover': 'Republican',
    '1933-Roosevelt': 'Democrat', '1937-Roosevelt': 'Democrat', '1941-Roosevelt': 'Democrat',
    '1945-Roosevelt': 'Democrat', '1949-Truman': 'Democrat', '1953-Eisenhower': 'Republican',
    '1957-Eisenhower': 'Republican', '1961-Kennedy': 'Democrat', '1965-Johnson': 'Democrat',
    '1969-Nixon': 'Republican', '1973-Nixon': 'Republican', '1977-Carter': 'Democrat',
    '1981-Reagan': 'Republican', '1985-Reagan': 'Republican', '1989-Bush': 'Republican',
    '1993-Clinton': 'Democrat', '1997-Clinton': 'Democrat', '2001-Bush': 'Republican',
    '2005-Bush': 'Republican', '2009-Obama': 'Democrat', '2013-Obama': 'Democrat',
    '2017-Trump': 'Republican', '2021-Biden': 'Democrat'
}

# Create a dictionary to store each speech's vocabulary
vocab_size = defaultdict(int)

# Process each inaugural speech
for fileid in inaugural.fileids():
    words = [word.lower() for word in inaugural.words(fileid)]
    unique_words = set(words)
    vocab_size[fileid] = len(unique_words)

# Create a DataFrame to associate speeches with their vocab size, president, year, and party
inaugural_data = []
for fileid, vocab in vocab_size.items():
    year, president = fileid.split('-')
    key = f"{year}-{president}"
    party = party_affiliations.get(key, 'Other')
    inaugural_data.append({
        'Year': int(year),
        'President': president,
        'Party': party,
        'Vocabulary_Size': vocab
    })

df_inaugural = pd.DataFrame(inaugural_data)
```
